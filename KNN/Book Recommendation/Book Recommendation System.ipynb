# Recommending Books based on Publishers

```python
df.publisher.value_counts()
```

```python
def recommend_books_publisher(x):
    a=df[df['publisher']== x][['title', 'average_rating']]  # From the filtered DataFrame, select only the 'title' and 'average_rating' columns
    a=a.sort_values(by ='average_rating' , ascending=False)
    return a.head(10)
```

```python
recommend_books_publisher('Vintage')
```

```python

recommend_books_publisher('Penguin Books')
```

```python
#select a publisher name from a dropdown list and then displays the top 10 books published by that selected publisher based on their average ratings.

@interact    #@interact decorator from the ipywidgets library to create an interactive function 
def recommend_books_publisher(publisher_name= list(df['publisher'].value_counts().index)):
    a = df[df['publisher'] == publisher_name][['title', 'average_rating']]
    a = a.sort_values(by = 'average_rating', ascending = False)
    return a.head(10)
```

# based upon Authors

```python
@interact
def recommend_books_authors(authors_name =list(df['authors'].value_counts().index)):
    a=df[df['authors']== authors_name][['title', 'average_rating']]
    a=a.sort_values(by='average_rating', ascending = False)
    return a.head(10)
```

# Recommending Books based on Language

```python
@interact
def recommend_books_lang(language=list(df['language_code'].value_counts().index)):
    a= df[df['language_code']== language][['title', 'average_rating']]
    a= a.sort_values(by='average_rating', ascending=False)
    return a.head(10)
```

# Data Preprocessing

```python
df.head()
```

```python
def num_to_obj(x):
    if x>0 and x<=1:
        return 'between 0 and 1'
    if x > 1 and x <= 2:
        return "between 1 and 2"
    if x > 2 and x <=3:
        return "between 2 and 3"
    if x >3 and x<=4:
        return "between 3 and 4"
    if x >4 and x<=5:
        return "between 4 and 5"
df['rating_obj'] = df['average_rating'].apply(num_to_obj)
```

```python
df['rating_obj'].value_counts()
```

```python
rating_df = pd.get_dummies(df['rating_obj'])
rating_df = rating_df.astype(int)
rating_df.head()
```

```python
language_df = pd.get_dummies(df['language_code'])
language_df = language_df.astype(int)
language_df.head()
```

```python
features = pd.concat([rating_df,language_df, df['average_rating'],
                    df['ratings_count'], df['title']], axis = 1)
features.set_index('title', inplace= True)
features.head()
```

```python
from sklearn.preprocessing import MinMaxScaler 
```

```python
scaler= MinMaxScaler()
```

```python
features_scaled = scaler.fit_transform(features)
features_scaled
```

```python
from sklearn import neighbors
```

```python
model= neighbors.NearestNeighbors(n_neighbors=5)
model.fit(features_scaled)
dist, idlist = model.kneighbors(features_scaled)
```

```python
@interact
def BookRecomender(book_name = list(df['title'].value_counts().index)):
    book_list_name = []
    book_id = df[df['title'] == book_name.index]
    book_id = book_id[0]
    for newid in idlist[book_id]:
        book_list_name.append(df.iloc[newid].title)
    return book_list_name
```
